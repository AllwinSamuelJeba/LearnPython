### Rock Paper Scissors Program


```python
#Importing random
import random

#Vars
rock ='''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''
paper = '''
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
'''

scissors ='''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)

'''

#Inputs

print("Enter one of the following option. ")
print("1 for" + rock)

print("2 for "+ paper)

print("3 for " + scissors)

userInput = int(input("Select 1, 2, or 3. \n"))

#Error messages in lines 71 - 72

computerSelection = random.randint(1,3)
print(computerSelection)

#Rules

if userInput == 1 and computerSelection == 1:
    print("You selected "+ rock + " and computer selected " + rock + " Hit run to try again.")
elif userInput == 2 and computerSelection == 2:
    print("You selected "+ paper+" and computer selected " + paper + " Hit run to try again.")
elif userInput == 3 and computerSelection == 3:
    print("You selected " + scissors + " and computer selected " + scissors + " Hit run to try again.")
elif userInput == 1 and computerSelection == 3:
    print("You selected "+ rock + " and computer selected " + scissors + " You win. ")
elif userInput == 1 and computerSelection == 2:
    print("You selected "+ rock + " and computer selected "+ paper+" Computer wins.")
elif userInput == 2 and computerSelection == 1:
    print("You selected "+ paper+" and computer selected " + rock + " You win.")
elif userInput == 2 and computerSelection == 3:
    print("You selected "+ paper+" and computer selected " + scissors + " Computer wins.")
elif userInput == 3 and computerSelection == 2:
    print("You selected " + scissors + " and computer selected "+ paper+" You win.")
elif userInput == 3 and computerSelection == 1:
    print("You selected Scissors and computer selected " + rock + " Computer wins.")
    
#Error

else:
    print(u"\u274c" + " Sorry. Input is not 1, 2, or 3. Hit run to try again.")
```
